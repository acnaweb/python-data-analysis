```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_pickle("datasets/pima-data.pkl")
```

```python
df.hist()
plt.show()
```

```python
# density plot
df.plot(kind = "density", subplots = True, layout = (3,3), sharex = False)
plt.show()
```

```python
# box plot
# age, skin e test com box no limite a valores menores
df.plot(kind = "box", subplots = True,  sharex = False)
plt.show()
```

```python
# age, skin e test com box no limite a valores menores
df[["age", "skin", "test"]].plot(kind = "box", layout = (3,1), subplots = True,  sharex = False)
plt.show()
```

```python
# correlacao
correlations = df.corr()

#plot da correlacao
import numpy as np

# criar figura
fig = plt.figure()

ax = fig.add_subplot(111)

# exibe a correlacao
cax = ax.matshow(correlations, vmin = -1, vmax = 1)

# barra lateral
fig.colorbar(cax)


# tamanho do quadrado
ticks = np.arange(0,9,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)

# nomes das colunas
ax.set_xticklabels(df.columns.values)
ax.set_yticklabels(df.columns.values)

# plot
plt.show()
```
