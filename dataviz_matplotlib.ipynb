```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_pickle("data/pima-data.pkl")
```

```python
df.hist()
plt.show()
```

```python
# density plot
df.plot(kind = "density", subplots = True, layout = (3,3), sharex = False)
plt.show()
```

```python
# box plot
# age, skin e test com box no limite a valores menores
df.plot(kind = "box", subplots = True,  sharex = False)
plt.show()
```

```python
# age, skin e test com box no limite a valores menores
df[["age", "skin", "test"]].plot(kind = "box", layout = (3,1), subplots = True,  sharex = False)
plt.show()
```

```python
# correlacao
correlations = df.corr()

#plot da correlacao
import numpy as np

# criar figura
fig = plt.figure()

ax = fig.add_subplot(111)

# exibe a correlacao
cax = ax.matshow(correlations, vmin = -1, vmax = 1)

# barra lateral
fig.colorbar(cax)


# tamanho do quadrado
ticks = np.arange(0,9,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)

# nomes das colunas
ax.set_xticklabels(df.columns.values)
ax.set_yticklabels(df.columns.values)

# plot
plt.show()
```

```python
# camadas sem dados
figure = plt.figure(figsize = (10, 8))
gs_master = mpl.gridspec.GridSpec(4,2, height_ratios=[1,2,8,2])

# camada 1 - title
gs_title = mpl.gridspec.GridSpecFromSubplotSpec(1,1, subplot_spec=gs_master[0, :])
title_axes = figure.add_subplot(gs_title[0])

# camada 2 - price
gs_price = mpl.gridspec.GridSpecFromSubplotSpec(1,1, subplot_spec=gs_master[1, :])
price_axes = figure.add_subplot(gs_price[0])

# camada 3 - risks
gs_risks1 = mpl.gridspec.GridSpecFromSubplotSpec(2,2, height_ratios=[2,1], subplot_spec=gs_master[2, :])
risks_and_loss_axes = figure.add_subplot(gs_risks1[0,:])
risks_axes = figure.add_subplot(gs_risks1[1,:1])
loss_axes = figure.add_subplot(gs_risks1[1:,1])

gs_risks2 = mpl.gridspec.GridSpecFromSubplotSpec(1,1, subplot_spec=gs_master[2, 1])
radar_axes = figure.add_subplot(gs_risks2[0])

gs_master.tight_layout(figure)
plt.show()
```
