```python
import pandas as pd
```

```python
ds1 = {"disciplina_id": ["1", "2", "3", "4", "5"],
       "nome": ["Bernardo", "Alan", "Mateus", "Ivo", "Gerson"],
       "sobrenome": ["Anderson", "Teixeira", "Amoedo", "Trindade", "Vargas"]
        }
```

```python
df_a = pd.DataFrame(ds1)
df_a
```

```python
ds2 = {"disciplina_id": ["4", "5", "6", "7", "8"],
       "nome": ["Roberto", "Mariana", "Ana", "Marcos", "Maria"],
       "sobrenome": ["Sampaio", "Fernandes", "Arantes", "Menezes", "Martins"]
        }
```

```python
df_b = pd.DataFrame(ds2)
df_b
```

```python
ds3 = {"disciplina_id": ["1", "2", "3", "4", "5", "7", "8", "9", "10", "11"],
      "teste_id": [81,75,75,71,76,84,95,61,57,90]}
```

```python
df_c = pd.DataFrame(ds3)
df_c
```

```python
database = {"df_a": df_a, "df_b": df_b, "df_c": df_c}
```

```python
# juncao vertical (normal)
df_alunos = pd.concat([df_a, df_b])
df_alunos 
```

```python
# juncao horizontal 
pd.concat([df_alunos, df_b], axis = 1)
```

```python
# join A e B
df_alunos_teste = pd.merge(df_alunos, df_c, on = "disciplina_id")
df_alunos_teste
```

```python
# outer join A e B
df_alunos_teste = pd.merge(df_alunos, df_c, on = "disciplina_id", how = "outer")
df_alunos_teste
```

```python
# inner join A e B
df_alunos_teste = pd.merge(df_alunos, df_c, on = "disciplina_id", how = "inner")
df_alunos_teste
```

```python
# left outer join A e B
df_alunos_teste = pd.merge(df_alunos, df_c, on = "disciplina_id", how = "left")
df_alunos_teste
```

```python
# right outer join A e B
df_alunos_teste = pd.merge(df_alunos, df_c, on = "disciplina_id", how = "right")
df_alunos_teste
```

```python
# right outer join A e B
df_alunos_teste = pd.merge(df_a, df_b, on = "disciplina_id", how = "left", suffixes = ("_left", "_right"))
df_alunos_teste
```

```python
# join baseado em index
pd.merge(df_a, df_b, right_index = True, left_index = True)
```
